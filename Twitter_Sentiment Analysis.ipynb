```python
all_words = []
for doc in documents_train:
    all_words += doc
```

```python
freq = nltk.FreqDist(all_words)
common = freq.most_common(8000)
features = [i[0] for i in common]
```

```python
features
```

```python
def get_feature_dict(words):
    current_features = {}
    words_set = set(words)
    for w in features:
        current_features[w] = w in words_set
    return current_features
```

```python
training_data = [get_feature_dict(doc) for doc in documents_train]
testing_data = [get_feature_dict(doc) for doc in documents_test]
```
